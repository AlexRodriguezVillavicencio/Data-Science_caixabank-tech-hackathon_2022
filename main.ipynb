## $$Carga-de-datos$$

$$***$$

```python
import pandas as pd
```

Trabajamos con el arhivo train.csv:

```python
df_train = pd.read_csv('./data/train.csv')
```

```python
df_train.columns
```

```python
df_train.isna()
```

Limpiamos los datos vacíos para que no interfieran en la predicción del modelo:

```python
new_df_train = df_train.dropna(axis=0)
```

```python
new_df_train.isna()
```

**********************************

## $$Interpretación$$

$$**$$

```python
new_df_train.describe()
```

```python
import plotly.express as px
```

Graficamos los datos para un mejor análisis:

```python
fig = px.scatter(new_df_train, x='Open', y='Adj Close', color='Volume')
fig.show()
```

```python
fig = px.scatter(new_df_train, x='Date', y='Adj Close', color='Volume')
fig.show()
```

**************************

## $$Modelo-predictivo$$

$$***$$

Definimos a 'Target' como nuestra variable dependiente:

```python
y = new_df_train.Target
```

Definimos nuestras varables independientes:

```python
data_features =['Open','Adj Close','Volume']
X = new_df_train[data_features]
```

```python
X.columns
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
forecast_model = DecisionTreeRegressor(random_state=1)
forecast_model.fit(X,y)
```

```python
forecast_model.predict(X.head())
```

***********************

## $$Validación-del-modelo$$

$$***$$

```python
from sklearn.metrics import mean_absolute_error
```

```python
predicted_model_target = forecast_model.predict(X)
```

```python
mean_absolute_error(y,predicted_model_target)
```

El error de nuestro modelo es 0. Entonces prodemos a validar el archivo test_x.csv:

```python
df_test = pd.read_csv('./data/test_x.csv')
```

```python
data_features2 = ['Open','Adj Close', 'Volume']
Xtest = df_test[data_features2]
Xtest.head()
```

```python
forecast_model.predict(Xtest.head())
```

```python
prediction =forecast_model.predict(Xtest)
prediction =prediction.astype(int)
prediction
```

```python
new_data = df_test[['test_index']]
new_data.head()
```

```python
export =new_data.assign(Target = prediction)
export.head()
```

*************************

## $$Exportación-de-datos$$

$$***$$

Guardamos el csv:

```python
export.to_csv('predictions.csv',header=True,index=False)
```

Transformamos el dataframe para luego exportarlo en un archivo json:

```python
export_json = export.set_index('test_index')
export_json = export_json.transpose()
```
